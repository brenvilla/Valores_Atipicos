### **Reemplazamiento de valores nulos**

```python
#Corroboramos valores nulos
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#Sustituci√≥n de valores nulos de la columna "subtotal_ventas_alimentos_bebidas"
#por el promedio

df["subtotal_ventas_alimentos_bebidas"]=df["subtotal_ventas_alimentos_bebidas"].fillna(round(df["subtotal_ventas_alimentos_bebidas"].mean(),1))  
```

```python
#Sustituci√≥n de los valores nulos de la columna "tarjetas_debito" por la mediana
#de la misma

df["tarjetas_debito"]=df["tarjetas_debito"].fillna(round(df["tarjetas_debito"].median(),1))
```

```python
#Sustituci√≥n de valores nulos de la columna "tarjetas_credito" por el valor 20

df["tarjetas_credito"]=df["tarjetas_credito"].fillna(20)
```

```python
#Sustituci√≥n de valores nulos de la columna "salon_ventas" por el string "Datos faltantes"

df["salon_ventas"]=df["salon_ventas"].fillna(0)
```

```python
#Sustituci√≥n de valores nulos de la columna "otros_medios" por el valor anterior

df["otros_medios"]=df["otros_medios"].fillna(method="ffill")
```

```python
#Sustituci√≥n de valores nulos de la columna "bebidas" por el valor posterior o que sigue

df["bebidas"]=df["bebidas"].fillna(method="bfill")
```

```python
##Sustituci√≥n de valores nulos de la columna "almacen" por el valor posterior o que sigue
df["almacen"]=df["almacen"].fillna(method="bfill")
```

```python
#Sustituci√≥n de valores nulos de la columna "panaderia" por el promedio
df["panaderia"]=df["panaderia"].fillna(round(df["panaderia"].mean(),1))  
```

```python
#Sustituci√≥n de valores nulos de la columa "lacteos" por la mediana
df["lacteos"]=df["lacteos"].fillna(round(df["lacteos"].median(),1))  
```

```python
#Sustituci√≥n de valores nulos de la columna "carnes" por el d√≠gito 30
df["carnes"]=df["carnes"].fillna(30)
```

```python
# Sustituci√≥n de valores nulos de la columna "verduleria_fruteria" por la expresi√≥n "00 00"
df["verduleria_fruteria"]=df["verduleria_fruteria"].fillna(0)
```

```python
#Sustituci√≥n de valores nulos de la columna "alimentos_preparados_rotiseria" por el valor anterior
df["alimentos_preparados_rotiseria"]=df["alimentos_preparados_rotiseria"].fillna(method="ffill")
```

```python
#Sustituci√≥n de valores nulos de la columna "indumentaria_calzado_textiles_hogar" por el valor posterior
df["indumentaria_calzado_textiles_hogar"]=df["indumentaria_calzado_textiles_hogar"].fillna(method="bfill")
```

```python
#Sustituci√≥n de valores nulos de la columna "electronicos_articulos_hogar" por el valor posterior
df["electronicos_articulos_hogar"]=df["electronicos_articulos_hogar"].fillna(method="bfill")
```

```python
#Sustituci√≥n de valores nulos de la columna "otros" por la cadena "--"
df["otros"]=df["otros"].fillna(0)
```

```python
#Comprobaci√≥n de valores nulos existentes
valores_nulos=df.isnull().sum()
valores_nulos
```

```python
#Total de valores nulos en todo el dataframe
valores_nulos=df.isnull().sum().sum()
valores_nulos
```

```python
VentasTotales_sin_nulos=df
VentasTotales_sin_nulos.to_csv("VentasTotales_sin_nulos.csv")
files.download("VentasTotales_sin_nulos.csv")
```

## **Eliminaci√≥n de Outliers por columna**

### **Variable 1: ventas_precios_corrientes**

```python
#M√©todo de desviaci√≥n estandar 
y=df["ventas_precios_corrientes"]
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers
outliers= df[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos limpios
data_clean= df[(y<=Limite_Superior)&(y>=Limite_Inferior)]
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean["ventas_precios_corrientes"]) 
plt.title("Outliers de ventas_precios_corrientes")
plt.show() #dibujamos el diagrama
```

```python
#M√©todo de IQR| Cuartiles 0.25 y 0.75 
y=df["ventas_precios_corrientes"]

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
#Encontramos Ouliers
outliers_iqr= df[(y>Limite_Superior_iqr)|(y<Limite_Inferior_iqr)]
outliers_iqr
```

```python
#Obtenemos datos limpios
data_clean_iqr= df[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean_iqr["ventas_precios_corrientes"]) 
plt.title("Outliers de ventas_precios_corrientes")
plt.show() #dibujamos el diagrama
```

```python
col_1=data_clean_iqr["ventas_precios_corrientes"]
col_1.to_csv("clean__iqr_ventas_precios_corrientes.csv")
files.download("clean__iqr_ventas_precios_corrientes.csv")
```
